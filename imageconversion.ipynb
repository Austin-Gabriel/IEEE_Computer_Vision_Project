```python
import os
from PIL import Image

source_dir = 'C:/Users/Blessing Austin/OneDrive - montclair.edu/Desktop/MSU/SETUP/PlantsImageClassification/data/HYDRO_POWER'
destination_dir = 'C:/Users/Blessing Austin/OneDrive - montclair.edu/Desktop/MSU/SETUP/PlantsImageClassification/data/HYDRO POWER'

def convert_images_to_png(source_folder, destination_folder):
    print(f"Checking {source_folder}...")
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
        print(f"Created directory {destination_folder}")

    filenames = os.listdir(source_folder)
    print(f"Files found: {filenames}")

    for filename in filenames:
        file_path = os.path.join(source_folder, filename)
        if os.path.isdir(file_path):
            new_destination = os.path.join(destination_folder, filename)
            convert_images_to_png(file_path, new_destination)
        else:
            dest_file_path = os.path.join(destination_folder, f"{os.path.splitext(filename)[0]}.png")
            try:
                with Image.open(file_path) as img:
                    img.convert('RGB').save(dest_file_path, 'PNG')
                    print(f"Converted and saved: {dest_file_path}")
            except Exception as e:
                print(f"Failed to convert {file_path}: {e}")

convert_images_to_png(source_dir, destination_dir)
```
